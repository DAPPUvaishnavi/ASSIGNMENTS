```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
```

```python
df = pd .read_csv("crime_data (1).csv")
```

```python
#data normalization
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)  
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=4, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
df['clust']=cluster_labels # creating a  new column and assigning it to new column 
df 
```

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
df1 = pd.read_csv("crime_data (1).csv")
```

```python
df1
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)  
```

```python
df_norm1 = norm_func(df1.iloc[:,1:])
df_norm1
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(df1.iloc[:,1:]))
data   
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm1)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(df_norm1)  
```

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    
```

```python
df2 = pd.read_csv("crime_data (1).csv")
```

```python
df2
```

```python
df2.info()
```

```python
df2 = df2.drop(['Unnamed: 0'],axis=1)    
```

```python
df2
```

```python
array=df2.values
array   
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X   
```

```python
# 1st method 
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(X)                      
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])   
```

```python
cl
```

```python
df2 = pd.concat([df2,cl],axis=1)  
df2    
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')   
```

```python
df1.plot(x="Murder",y ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')      
```

```python
dl = dbscan.labels_     
```

```python
import sklearn
sklearn.metrics.silhouette_score(X, dl)   
```

```python
#method 2 nearest neighbour for elbow curve
```

```python
from sklearn.neighbors import NearestNeighbors  
```

```python
# n = 2n
```

```python
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)  # find all 8 distance
```

```python
# sort distance in ascending order and plot distances
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)  
```
